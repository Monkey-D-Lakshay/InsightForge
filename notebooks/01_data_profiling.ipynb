```python
# Import required libraries for data analysis and file handling

import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Define the path to the folder containing the raw datasets
RAW_DATA_PATH = Path("../data/raw")

# Display the path to verify it is correct
RAW_DATA_PATH
```

```python
#Python automatically finds every CSV in the raw folder.
csv_files = sorted(RAW_DATA_PATH.glob("*.csv"))

# Print the total number of CSV files found
print(f"Found {len(csv_files)} CSV files:\n")

# Print the name of each CSV file
for file in csv_files:
    print(file.name)
```

```python
# Load all CSV files into a dictionary of DataFrames for easy access

datasets = {}

for file in csv_files:
    dataset_name = file.stem
    datasets[dataset_name] = pd.read_csv(file)

print(f"Successfully loaded {len(datasets)} datasets.")
print("\nAvailable datasets:")

for name in datasets:
    print(f"- {name}")
```

```python
# Display basic information about each dataset

for name, df in datasets.items():
    print("=" * 70)
    print(f"Dataset: {name}")
    print("=" * 70)

    print(f"Rows, Columns : {df.shape}")
    print(f"\nColumns:\n{list(df.columns)}")

    print("\nData Types:")
    print(df.dtypes)

    print("\nFirst 5 Rows:")
    display(df.head())

    print("\n")
```

```python
# Generate a summary of data quality for every dataset

summary = []

for name, df in datasets.items():
    summary.append({
        "Dataset": name,
        "Rows": df.shape[0],
        "Columns": df.shape[1],
        "Missing Values": df.isnull().sum().sum(),
        "Duplicate Rows": df.duplicated().sum()
    })

summary_df = pd.DataFrame(summary)

summary_df.sort_values(by="Rows", ascending=False)
```

```python
# Display missing values column-wise for each dataset

for name, df in datasets.items():
    missing = df.isnull().sum()
    missing = missing[missing > 0]

    if not missing.empty:
        print("=" * 70)
        print(f"Dataset: {name}")
        print("=" * 70)
        print(missing)
        print()
```

```python
# Display ID columns and their uniqueness for each dataset

for name, df in datasets.items():
    print("=" * 70)
    print(f"Dataset: {name}")
    print("=" * 70)

    id_columns = [col for col in df.columns if col.endswith("_id")]

    for col in id_columns:
        print(f"{col:<30} Unique Values: {df[col].nunique()}")

    print()
```
