# 5. Star Schema Design

## Objective

The objective of this section is to identify the business process, determine the fact table and dimension tables, and design a Star Schema for the Olist e-commerce data warehouse.

```python
# Define the proposed Star Schema for the data warehouse

FACT_TABLE = "Fact_Order_Items"

DIMENSION_TABLES = [
    "Dim_Customers",
    "Dim_Products",
    "Dim_Sellers",
    "Dim_Date",
    "Dim_Payments"
]

print(f"Fact Table: {FACT_TABLE}")

print("\nDimension Tables:")
for table in DIMENSION_TABLES:
    print(f"- {table}")
```

## Business Process

Before designing a data warehouse, we must identify the business process that generates value.

In the Olist e-commerce platform, the core business process is the sale of products through customer orders. Since a single order can contain multiple products from different sellers, each product sold represents a separate business transaction.

Therefore, the warehouse will be designed at the **order item level**, where each row in the fact table represents one product sold in one order.

## Why is `order_items` the Fact Table?

Although customers place orders, each order may contain multiple products sold by different sellers.

Using the `order_items` dataset as the fact table allows us to analyze:

- Revenue by product
- Revenue by seller
- Revenue by category
- Average selling price
- Freight cost
- Product-level sales trends

This provides a much more detailed and flexible analytical model than using the orders table alone.

## Designing the Star Schema

After identifying the business process and selecting the fact table, the next step is to design the structure of each table.

A well-designed Star Schema separates transactional data (facts) from descriptive information (dimensions). This improves query performance, reduces redundancy, and makes analytical reporting much simpler.

## Fact Table Measures

Measures are numeric values that can be aggregated to produce business insights.

The Fact_Order_Items table contains two primary measures:

- Price
- Freight Value

These measures will be used to calculate revenue, shipping costs, average order value, and other key performance indicators.

## Fact Table Structure

The Fact_Order_Items table stores each product sold within an order.

Each row represents one business transaction and contains:
- Foreign keys to related dimension tables.
- Business measures used for analytics.
- A surrogate primary key for efficient warehouse management.

| Column Name | Type | Description |
|--------------|------|-------------|
| order_item_key | Surrogate PK | Unique warehouse identifier |
| customer_key | FK | Links to Customer Dimension |
| product_key | FK | Links to Product Dimension |
| seller_key | FK | Links to Seller Dimension |
| date_key | FK | Links to Date Dimension |
| payment_key | FK | Links to Payment Dimension |
| order_id | Degenerate Dimension | Original business order ID |
| price | Measure | Product selling price |
| freight_value | Measure | Shipping cost |

# Customer Dimension (Dim_Customers)

## Objective

The Customer Dimension stores descriptive information about customers.

It enables business users to analyze sales based on customer location and demographics without storing repetitive customer details inside the fact table.

Each customer appears only once in this dimension and is referenced by the fact table using a surrogate key.

| Column Name | Type | Description |
|--------------|------|-------------|
| customer_key | Surrogate PK | Warehouse primary key |
| customer_id | Business Key | Original customer identifier |
| customer_unique_id | Business Attribute | Unique customer across multiple orders |
| customer_zip_code_prefix | Attribute | Customer ZIP code |
| customer_city | Attribute | Customer city |
| customer_state | Attribute | Customer state |

# Product Dimension (Dim_Products)

## Objective

The Product Dimension stores descriptive information about products.

It allows analysts to evaluate sales performance by product category, size, weight, and other characteristics while avoiding duplication of product information in the fact table.

| Column Name | Type | Description |
|--------------|------|-------------|
| product_key | Surrogate PK | Warehouse primary key |
| product_id | Business Key | Original product identifier |
| product_category_name | Attribute | Product category |
| product_name_length | Attribute | Product name length |
| product_description_length | Attribute | Product description length |
| product_photos_qty | Attribute | Number of product images |
| product_weight_g | Attribute | Product weight |
| product_length_cm | Attribute | Product length |
| product_height_cm | Attribute | Product height |
| product_width_cm | Attribute | Product width |

# Seller Dimension (Dim_Sellers)

## Objective

The Seller Dimension contains information about sellers participating in the marketplace.

It enables analysis of sales performance by seller and geographic region.

| Column Name | Type | Description |
|--------------|------|-------------|
| seller_key | Surrogate PK | Warehouse primary key |
| seller_id | Business Key | Original seller identifier |
| seller_zip_code_prefix | Attribute | Seller ZIP code |
| seller_city | Attribute | Seller city |
| seller_state | Attribute | Seller state |

# Date Dimension (Dim_Date)

## Objective

The Date Dimension provides a centralized calendar for time-based analysis.

Instead of storing only a purchase date, the warehouse stores additional attributes such as day, month, quarter, year, weekday, and weekend indicators. This simplifies reporting and improves query performance for time-based analytics.

| Column Name | Type | Description |
|--------------|------|-------------|
| date_key | Surrogate PK | Warehouse primary key |
| full_date | Date | Calendar date |
| day | Attribute | Day of month |
| month | Attribute | Month number |
| month_name | Attribute | Month name |
| quarter | Attribute | Quarter of the year |
| year | Attribute | Calendar year |
| day_of_week | Attribute | Day name |
| is_weekend | Attribute | Weekend flag |

# Payment Dimension (Dim_Payments)

## Objective

The Payment Dimension stores payment-related information for customer transactions. It enables analysis of payment methods, installment behavior, and payment trends across the business.

| Column Name | Type | Description |
|--------------|------|-------------|
| payment_key | Surrogate PK | Warehouse primary key |
| payment_type | Attribute | Credit Card, Debit Card, Voucher, etc. |
| payment_installments | Attribute | Number of installments |
